# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f1 = x * sp.exp(-x**2)
f2 = x + 1

x_vals = np.linspace(-1, 3, 100)
y_vals_f1 = [sp.lambdify(x, f1)(val) for val in x_vals]
y_vals_f2 = [sp.lambdify(x, f2)(val) for val in x_vals]

f = sp.lambdify(x, f1 - f2)
intersections = np.roots(f(x_vals))
area = np.trapz(np.minimum(y_vals_f1, y_vals_f2), x_vals)

plt.plot(x_vals, y_vals_f1, label='y = xe^(-x^2)')
plt.plot(x_vals, y_vals_f2, label='y = x + 1')
plt.fill_between(x_vals, y_vals_f1, y_vals_f2, where=(x_vals >= intersections[0]) & (x_vals <= intersections[1]), color='yellow', alpha=0.3)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Área entre as curvas')
plt.legend()
plt.grid(True)
plt.show()

print('Área entre as curvas:', area)
```

A)

Para resolver a integral

$$\int \cos(3t) \sin(8t) dt,\$$

usando a identidade trigonométrica:


$$\sin(A)\cos(B) = \frac{1}{2}[\sin(A-B) + \sin(A+B)$$

Aplicando essa identidade aintegral:


$$\int \cos(3t) \sin(8t) dt = \frac{1}{2} \int [\sin(8t+3t) + \sin(8t-3t)] dt.$$

Simplificando os argumentos dos senos:


$$\int \cos(3t) \sin(8t) dt = \frac{1}{2} \int [\sin(11t) + \sin(5t)] dt.$$

Resolvendo as duas integrais separadamente.

Em $$\int \sin(11t) dt$$. Utilizando a regra de integração do seno:


$$\int \sin(11t) dt = -\frac{1}{11} \cos(11t) + C_1$,

onde $C_1$ é a constante de integração.

Resolve-se $$\int \sin(5t) dt$$ usando a mesma regra:


$$\int \sin(5t) dt = -\frac{1}{5} \cos(5t) + C_2$$,

onde $C_2$ é outra constante de integração.

Ao voltar a integral original:


$$\int \cos(3t) \sin(8t) dt = \frac{1}{2} \int [\sin(11t) + \sin(5t)] dt.$$

Substituindo as integrais:


$$\int \cos(3t) \sin(8t) dt = \frac{1}{2} \left(-\frac{1}{11} \cos(11t) - \frac{1}{5} \cos(5t) \right) + C$$,

onde $$C = \frac{1}{2}(C_1 + C_2)$$ é a constante de integração final.

A solução da integral é:


$$\int \cos(3t) \sin(8t) dt = -\frac{1}{22} \cos(11t) - \frac{1}{10} \cos(5t) + C$$,

onde $C$ é uma constante arbitrária.


B)

Para resolver a integral 

$$\int 4x \cos(2-3x) dx,$$

A fórmula de integração por partes é dada por:

$$\int u dv = uv - \int v du,$$

onde $u$ e $v$ são funções diferenciáveis em relação a $x$.

Foi escolhido $u = 4x$ e $dv = \cos(2-3x) dx$.

É necessário calcular $du$ e $v$. 

Para $du$, toma-se a derivada de $u$ em relação a $x$:

$$du = \frac{d}{dx}(4x) dx = 4 dx.$$

Para $v$, integramos $dv$ em relação a $x$:

$$v = \int \cos(2-3x) dx.$$

Substituindo esses valores na fórmula de integração:

$$\int 4x \cos(2-3x) dx = 4x \int \cos(2-3x) dx - \int \left(\int \cos(2-3x) dx \right) 4 dx.$$

Simplificando:

$$\int 4x \cos(2-3x) dx = 4x \int \cos(2-3x) dx - 4 \int \left(\int \cos(2-3x) dx \right) dx.$$

Ao resolver a primeira integral $\int \cos(2-3x) dx$. É feita uma substituição de variável, $u = 2-3x$, e se obtem $du = -3 dx$, ou seja, $dx = -\frac{1}{3} du$. Assim:

$$\int \cos(2-3x) dx = \int \cos(u) \left(-\frac{1}{3} du\right) = -\frac{1}{3} \int \cos(u) du = -\frac{1}{3} \sin(u) + C_1,$$

onde $C_1$ é a constante de integração.

Substituíndo essa integral de volta na expressão original:

$$\int 4x \cos(2-3x) dx = 4x \left(-\frac{1}{3} \sin(u) + C_1\right) - 4 \int \left(-\frac{1}{3} \sin(u) + C_1\right) dx.$$

Simplificando:

$$\int 4x \cos(2-3x) dx = -\frac{4}{3} x \sin(u) + 4C_1x - \frac{4}{3} \int \sin(u) dx + 4C_2,$$

onde $C_2$ é outra constante de integração.

Ao resolver a segunda integral $\int \sin(u) dx$. É feita uma substituição de variável, $v = 2-3x$, onde se obtem $dv = -3 dx$, ou seja, $dx = -\frac{1}{3} dv$. Temos então:

$$\int \sin(u) dx = \int \sin(u) \left(-\frac{1}{3} dv\right) = -\frac{1}{3} \int \sin(u) dv = -\frac{1}{3} (-\cos(v)) + C_3,$$

onde $C_3$ é mais uma constante de integração.

Substituindo essa integral na expressão anterior:

$$\int 4x \cos(2-3x) dx = -\frac{4}{3} x \sin(2-3x) + 4C_1x - \frac{4}{3} \left(-\frac{1}{3} (-\cos(2-3x)) + C_3\right) + 4C_2.$$

Simplificando e reescrevendo em termos da variável original $x$:

$$\int 4x \cos(2-3x) dx = -\frac{4}{3} x \sin(2-3x) - \frac{4}{9} \cos(2-3x) + 4C_1x + \frac{4}{9}C_3 + 4C_2.$$

A solução da integral é:

$$\int 4x \cos(2-3x) dx = -\frac{4}{3} x \sin(2-3x) - \frac{4}{9} \cos(2-3x) + 4C_1x + \frac{4}{9}C_3 + 4C_2,$$

onde $C_1$, $C_2$ e $C_3$ são constantes arbitrárias.


# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração. 

##a)
$$\int_0^2 x^2+1 \space dx$$

##b) 
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



```python
import numpy as np
import matplotlib.pyplot as plt

# Questão a)
x_a = np.linspace(0, 2, 100)  # Valores de x entre 0 e 2
y_a = x_a**2 + 1              # Função a ser integrada

# Integral definida
integral_a = np.trapz(y_a, x_a)

# Gráfico
plt.plot(x_a, y_a, label='Função a ser integrada')
plt.fill_between(x_a, y_a, alpha=0.2)
plt.axvline(x=0, color='r', linestyle='--')
plt.axvline(x=2, color='r', linestyle='--')
plt.title(f'Gráfico da função a), integral = {integral_a:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Questão b)
x_b = np.linspace(-3, 1, 100)  # Valores de x entre -3 e 1
y_b = 6 * x_b**2 - 5 * x_b + 2  # Função a ser integrada

# Integral definida
integral_b = np.trapz(y_b, x_b)

# Gráfico
plt.plot(x_b, y_b, label='Função a ser integrada')
plt.fill_between(x_b, y_b, alpha=0.2)
plt.axvline(x=-3, color='r', linestyle='--')
plt.axvline(x=1, color='r', linestyle='--')
plt.title(f'Gráfico da função b), integral = {integral_b:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Encontre o valor da constante C para a antiderivada $F(x)$ da função $f(x) = 3x^3 - 2x + 1$ de tal forma que $F(1)=0$. Apresente seus cálculos em latex e explique as decisões que tomou. 

É necessário encontrar primeiro a antiderivada $F(x)$ da função $f(x)$ dada.

$$f(x) = 3x^3 - 2x + 1$$

Para encontrar a antiderivada, é integrada a função $f(x)$ em relação a $x$.

$$F(x) = \int (3x^3 - 2x + 1) dx$$

Integrando termo a termo:

$$F(x) = \frac{3}{4}x^4 - x^2 + x + C$$

Para encontrar o valor de $C$ que torna a equação $F(1) = 0$ verdadeira, substituímos $x=1$ na equação acima:

$$F(1) = \frac{3}{4}(1)^4 - (1)^2 + (1) + C$$
$$0 = \frac{3}{4} - 1 + 1 + C$$
$$C = \frac{1}{4}$$

O valor da constante $C$ que torna a equação $F(1) = 0$ verdadeira é $\frac{1}{4}$.
