# Questão 1
Uma doença se espalha de forma exponencial, com uma taxa de contágio de 0,05 por dia. Se inicialmente 1000 pessoas são infectadas, faça um gráfico que mostre como o número de pessoas infectadas varia com o tempo e determine quantas pessoas estarão infectadas após 30 dias. Utilize a equação $I = I_0 * e^{rt}$, onde $I_0$ é o número inicial de infectados, `r` é a taxa de contágio diária (em decimal) e $t$ é o tempo decorrido em dias.



```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 31, 1)
I = 1000 * np.exp(0.05*t)

plt.plot(t, I)
plt.xlabel('Tempo (dias)')
plt.ylabel('Infectados')
plt.title('Propagação de doença')
plt.show()

```

# Questão 2
Usando, a seu critério, as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $f(x) = 3x^2 - 6^x +48$

2. $f(x) = \log_{10}(x^2 - 3x) - \log_{10}(x - 3)$

#Resposta 

\begin{aligned}
x &= \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} \
&= \frac{-(-6) \pm \sqrt{(-6)^2 - 4(3)(48)}}{2(3)} \
&= \frac{6 \pm \sqrt{36 - 576}}{6} \
&= \frac{6 \pm \sqrt{-540}}{6} \
&= \frac{6 \pm 6i\sqrt{15}}{6} \
&= 1 \pm i\sqrt{15}
\end{aligned}

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 3*x**2 - 6*x + 48

x = np.linspace(-10, 10, 1000)
y = f(x)

plt.plot(x, y)
plt.ylabel('y')
plt.xlabel('Crescente para x>1, e decrescente para x<1')
plt.title('Gráfico da função')
plt.grid(True)
plt.show()
```

# Questão 3
Usando o Sympy, ou qualquer outro recurso disponível na internet, escreva, com qualidade matemática o passo-a-passo algébrico que leva a determinação dos seguintes limites. Ao final, para cada uma das equações você deverá traçar um gráfico da função. 

1. $\lim_{x \to 1} \frac{x^{10} - 1}{x^5 - 1}$

2. $\lim_{x \to \pi} \frac{\cos(x) - 1}{x - \pi}$

#Resposta

1. $\lim_{x \to 1} \frac{x^{10} - 1}{x^5 - 1}$

A função é:

$\lim_{x \to 1} (\frac{x^{10} - 1}{5^x - 1})$

Fazendo fatoração:

$\lim_{x \to 1} {((x+1)(x^4 - x^3 - x^2 - x +1))}$ 


Aplicando o limite que temos:

${((1) + 1)((1)^4 - (1)^3 + (1)^2 1 (1) +1)}$

A resposta final é 2


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (x**10 - 1) / (x**5 - 1)

x = np.linspace(0.9, 1.1, 1000)
y = f(x)

plt.plot(x, y)
plt.ylim(0, 30)
plt.xlim(0.9, 1.1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função')
plt.grid(True)
plt.show()
```

2. $lim_{x \to \pi} \frac{\cos(x) - 1}{x - \pi}$

A função é:

$\frac{\cos(x) - 1}{x - \pi}$

$lim_{x \to \pi} (\frac{\cos(x) - 1}{x - \pi})$

Aplicando o limite:

$\frac{\cos((\pi)) - 1}{(\pi) - \pi}$

∞

Resposta final: ∞

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (np.cos(x) - 1) / (x - np.pi)

x = np.linspace(np.pi - 0.1, np.pi + 0.1, 1000)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da função (cos(x) - 1) / (x - pi) quando x se aproxima de pi')
plt.grid(True)
plt.show()
```
